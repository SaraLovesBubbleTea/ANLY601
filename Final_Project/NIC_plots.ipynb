```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from NIC import NIC
```

```python
# function generate Gaussian Ring
def gr(n, w, r, mean=[0, 0], cov=[[1, 0], [0, 1]]):
    """
    n: number of points
    w: width of the ring
    r: radius
    mean: default mean [0,0]
    cov: default cov [[1,0],[0,1]]
    """
    x, y = np.random.multivariate_normal(mean, cov, n).T
    xy = zip(x, y)
    result = []
    for z in xy:
        z = np.array(z)
        result.append(z / w + r * z / np.linalg.norm(z))

    return result
```

```python
# function generate Gaussian distribution
def gd(n, mean = [0,0], cov = [[1,0],[0,1]],skew = [0,0]):
    X  = np.random.multivariate_normal(mean, cov, n) + skew
    return X
```

```python
# Basic plot comparision
X1 = gd(500)
X2 = gd(500,skew=[3,3])
```

```python
X = np.concatenate((X1,X2),axis = 0)
df = pd.DataFrame(X,columns=['X','Y'])
```

```python
sns.scatterplot(df['X'],df['Y'])
```

```python
kmean = KMeans(n_clusters=2)
```

```python
kmean.fit(X)
```

```python
df['k_label'] = kmean.labels_
```

```python
sns.scatterplot(df['X'],df['Y'],hue = df['k_label'])
```

```python
t = NIC()
t.fit(X,3)
```

```python
df['label'] = t.labels_
```

```python
sns.scatterplot(df['X'],df['Y'],hue = df['label'])
```
